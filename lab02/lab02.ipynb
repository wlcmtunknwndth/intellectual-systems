```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
```

```python
housing = fetch_california_housing()

df_housing = pd.DataFrame(housing.data, columns=housing.feature_names)
df_housing['target'] = housing.target

df_housing.head()
```

## 2. Выполняем стандартизацию данных (при необходимости).

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = housing.data
y = housing.target

# Выполняем стандартизацию признаков
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

df_scaled = pd.DataFrame(X_scaled, columns=housing.feature_names)
df_scaled['target'] = y

df_scaled.head()
```

## 3. Выполняем расширение базовых признаков за счет применения полиномиальной модели (степень = 3)

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly.fit_transform(X_scaled)

feature_names = poly.get_feature_names_out(housing.feature_names)

df_poly = pd.DataFrame(X_poly, columns=feature_names)
df_poly['target'] = y

df_poly.iloc[:, :10].head()
```

## 4. Разбиваем датасеты (базовый и расширенный) на 2 множества (обучающее и тестовое).

```python
from sklearn.model_selection import train_test_split
```

```python
# Разбиваем базовый датасет (стандартизированные признаки)
X_train_basic, X_test_basic, y_train_basic, y_test_basic = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42
)

# Разбиваем расширенный датасет (полиномиальные признаки)
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(
    X_poly, y, test_size=0.3, random_state=42
)

print("Базовый датасет (стандартизированные признаки):")
print(f"  Обучающая выборка: {X_train_basic.shape[0]} объектов, {X_train_basic.shape[1]} признаков")
print(f"  Тестовая выборка: {X_test_basic.shape[0]} объектов, {X_test_basic.shape[1]} признаков")

print("\nРасширенный датасет (полиномиальные признаки):")
print(f"  Обучающая выборка: {X_train_poly.shape[0]} объектов, {X_train_poly.shape[1]} признаков")
print(f"  Тестовая выборка: {X_test_poly.shape[0]} объектов, {X_test_poly.shape[1]} признаков")

print("\nЦелевая переменная (одинаковая для обоих датасетов):")
print(f"  Обучающая выборка: {y_train_basic.shape[0]} значений")
print(f"  Тестовая выборка: {y_test_basic.shape[0]} значений")
```

## 5. Выполняем обучение линейной и полиномиальной моделей аналитическим методом - не используя готовое библиотечное решение.

```python
# Обучение линейной модели с использованием нормальных уравнений: w = (X^T * X)^(-1) * X^T * y
def train_linear_model_analytical(X, y):
    # Добавляем столбец единиц для cвободного члена
    X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
    
    # Вычисляем веса по нормальным уравнениям
    XTX = X_with_intercept.T @ X_with_intercept
    XTy = X_with_intercept.T @ y
    
    # Решаем систему уравнений
    try:
        w = np.linalg.solve(XTX, XTy)
    except np.linalg.LinAlgError:
        # Если матрица вырождена, используем псевдообратную
        w = np.linalg.pinv(XTX) @ XTy
    
    return w
```

```python
def predict_linear_model(X, w):
    """Предсказание с использованием обученной линейной модели"""
    X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
    return X_with_intercept @ w
```

```python
# Обучение линейной модели на базовых признаках
w_basic = train_linear_model_analytical(X_train_basic, y_train_basic)
print(f"Обучено весов: {len(w_basic)} (включая intercept)")

# Обучение линейной модели на полиномиальных признаках
w_poly = train_linear_model_analytical(X_train_poly, y_train_poly)
print(f"Обучено весов: {len(w_poly)} (включая intercept)")

# Предсказания на тестовых данных
y_pred_basic = predict_linear_model(X_test_basic, w_basic)
y_pred_poly = predict_linear_model(X_test_poly, w_poly)

print(f"Линейная модель (базовые признаки): {len(w_basic)} параметров")
print(f"Линейная модель (полиномиальные признаки): {len(w_poly)} параметров")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def calculate_regression_metrics(y_true, y_pred, model_name):
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    rmse = np.sqrt(mse)
    
    print(f"\n{model_name}:")
    print(f"  MSE (Mean Squared Error): {mse:.4f}")
    print(f"  RMSE (Root Mean Squared Error): {rmse:.4f}")
    print(f"  MAE (Mean Absolute Error): {mae:.4f}")
    print(f"  R² (Coefficient of Determination): {r2:.4f}")
    
    return {'mse': mse, 'rmse': rmse, 'mae': mae, 'r2': r2}

# Вычисляем метрики для обеих моделей
metrics_basic = calculate_regression_metrics(y_test_basic, y_pred_basic, 
                                           "Линейная модель (базовые признаки)")
metrics_poly = calculate_regression_metrics(y_test_poly, y_pred_poly, 
                                          "Линейная модель (полиномиальные признаки)")
```

## 6. Выполняем обучение линейной и полиномиальной моделей градиентным спуском - не используя готовое библиотечное решение.

```python
# Функция для обучения линейной модели градиентным спуском
def train_linear_model_gradient_descent(X, y, learning_rate=0.001, n_iterations=1000, tolerance=1e-6):
    X_with_intercept = np.column_stack([np.ones(X.shape[0]), X])
    w = np.random.randn(X_with_intercept.shape[1]) * 0.01
    losses = []
    
    for i in range(n_iterations):
        y_pred = X_with_intercept @ w
        error = y_pred - y
        loss = np.mean(error ** 2)
        losses.append(loss)
        
        gradient = (2 / len(y)) * (X_with_intercept.T @ error)
        
        # Gradient clipping
        grad_norm = np.linalg.norm(gradient)
        if grad_norm > 1.0:
            gradient = gradient / grad_norm
        
        w_new = w - learning_rate * gradient
        
        # Проверка сходимости весов
        if np.linalg.norm(w_new - w) < tolerance:
            print(f"  Сходимость достигнута на итерации {i+1}")
            break
            
        # Проверка расходимости
        if not np.isfinite(loss) or not np.isfinite(grad_norm):
            print(f"  Расходимость на итерации {i+1}")
            break
            
        # Проверка сходимости по loss
        if i > 0 and abs(losses[-1] - losses[-2]) < tolerance:
            print(f"  Сходимость по loss на итерации {i+1}")
            break
        
        w = w_new
        
        if i % 100 == 0:
            print(f"  Итерация {i}, Loss: {loss:.4f}")
    
    return w, losses

print("Обучение линейной модели на базовых признаках с градиентным спуском)...")
w_basic_gd, losses_basic = train_linear_model_gradient_descent(X_train_basic, y_train_basic, learning_rate=0.001)

print("Обучение линейной модели на полиномиальных признаках с градиентным спуском...")
w_poly_gd, losses_poly = train_linear_model_gradient_descent(X_train_poly, y_train_poly, learning_rate=0.0001)

# Предсказания на тестовых данных
y_pred_basic_gd = predict_linear_model(X_test_basic, w_basic_gd)
y_pred_poly_gd = predict_linear_model(X_test_poly, w_poly_gd)
```

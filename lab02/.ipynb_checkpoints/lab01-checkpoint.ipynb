## 2. Выполняем стандартизацию данных.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X[:5], X_scaled[:5]
```

## 3. Разбиваем выборку на 2 множества (обучающее и тестовое).

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)
```

```python
print("Расчет ядер классов:")

centroids = []
for class_label in np.unique(y_train):
    # Выбираем объекты текущего класса
    class_samples = X_train[y_train == class_label]
    
    # Рассчитываем ядро класса
    centroid = np.mean(class_samples, axis=0)
    centroids.append(centroid)
    
    print(f"\nКласс {class_label} ({target_names[class_label]}):")
    print(f"  Количество объектов: {len(class_samples)}")
    print(f"  Центроид (первые 5 компонент): {centroid[:5]}")
    print(f"  Размер центроида: {centroid.shape}")

centroids = np.array(centroids)
print(f"\nМатрица ядер классов: {centroids.shape}")
```

```python
print("\nРасчет матриц ковариации для каждого класса:")

covariance_matrices = []
for class_label in np.unique(y_train):
    # Выбираем объекты текущего класса
    class_samples = X_train[y_train == class_label]
    
    # Рассчитываем матрицу ковариации
    cov_matrix = np.cov(class_samples, rowvar=False)
    covariance_matrices.append(cov_matrix)
    
    print(f"\nКласс {class_label} ({class_names[class_label]}):")
    print(f"  Размер матрицы ковариации: {cov_matrix.shape}")
    print(f"  След матрицы: {np.trace(cov_matrix):.4f}")
    print(f"  Определитель матрицы: {np.linalg.det(cov_matrix):.4f}")
    
    # Диагональные элементы (дисперсии признаков)
    variances = np.diag(cov_matrix)
    print(f"  Диапазон дисперсий: [{np.min(variances):.4f}, {np.max(variances):.4f}]")

covariance_matrices = np.array(covariance_matrices)
covariance_matrices.shape
```

## 5. Для каждого объекта тестовой выборки измеряем расстояние до ядер (по метрике Евклида и по метрике Махаланобиса) и присваиваем метку класса.

```python
from scipy.spatial.distance import cdist
from sklearn.metrics import accuracy_score
```

```python
def calculate_class_covariances(X, y):
    classes = np.unique(y)
    covariances = {}
    means = {}
    
    for cls in classes:
        X_cls = X[y == cls]
        means[cls] = np.mean(X_cls, axis=0)
        covariances[cls] = np.cov(X_cls, rowvar=False)
        
    return means, covariances
```

```python

# Функция для классификации с использованием метрики Евклида
def euclidean_kernel_classification(X_train, y_train, X_test, gamma=1.0):
    n_train = X_train.shape[0]
    n_test = X_test.shape[0]
    n_classes = len(np.unique(y_train))
    
    # Вычисляем матрицу расстояний между тестовыми и обучающими объектами
    distances = cdist(X_test, X_train, metric='euclidean')
    
    # Применяем гауссово ядро
    kernel_matrix = np.exp(-gamma * distances**2)
    
    # Предсказываем классы
    y_pred = []
    
    for i in range(n_test):
        class_scores = np.zeros(n_classes)
        
        for cls in range(n_classes):
            # Находим индексы объектов данного класса в обучающей выборке
            class_indices = (y_train == cls)
            
            if np.sum(class_indices) > 0:
                # Суммируем значения ядра для объектов данного класса
                class_scores[cls] = np.sum(kernel_matrix[i, class_indices])
        
        # Выбираем класс с максимальной суммой
        y_pred.append(np.argmax(class_scores))
    
    return np.array(y_pred)

# Функция для классификации с использованием метрики Махаланобиса
def mahalanobis_kernel_classification(X_train, y_train, X_test, class_means, class_covariances, gamma=1.0):
    n_test = X_test.shape[0]
    classes = list(class_means.keys())
    
    y_pred = []
    
    for i in range(n_test):
        x_test = X_test[i]
        class_scores = []
        
        for cls in classes:
            # Вычисляем расстояние Махаланобиса
            diff = x_test - class_means[cls]
            
            # Регуляризация ковариационной матрицы для устойчивости
            cov_matrix = class_covariances[cls]
            # Добавляем небольшое значение к диагонали для предотвращения вырожденности
            reg_cov = cov_matrix + np.eye(cov_matrix.shape[0]) * 1e-6
            
            try:
                # Вычисляем расстояние Махаланобиса
                mahal_dist = diff @ np.linalg.inv(reg_cov) @ diff.T
                
                # Применяем гауссово ядро
                kernel_value = np.exp(-gamma * mahal_dist)
                class_scores.append((kernel_value, cls))
                
            except np.linalg.LinAlgError:
                # В случае проблем с матрицей, используем большое расстояние
                class_scores.append((0, cls))
        
        # Выбираем класс с максимальным значением ядра
        best_class = max(class_scores, key=lambda x: x[0])[1]
        y_pred.append(best_class)
    
    return np.array(y_pred)
```

```python
class_means, class_covariances = calculate_class_covariances(X_train, y_train)

y_pred_euclidean = euclidean_kernel_classification(X_train, y_train, X_test)

y_pred_mahalanobis = mahalanobis_kernel_classification(
    X_train, y_train, X_test, class_means, class_covariances
)

accuracy_euclidean = accuracy_score(y_test, y_pred_euclidean)
accuracy_mahalanobis = accuracy_score(y_test, y_pred_mahalanobis)

print(f"\nТочность классификации (Евклидово расстояние): {accuracy_euclidean:.4f}")
print(f"Точность классификации (Махаланобис): {accuracy_mahalanobis:.4f}")
```

## 6. Рассчитываем метрики качества проведенной классификации.

```python
from sklearn.metrics import (accuracy_score, precision_score, recall_score, 
                            f1_score, confusion_matrix, classification_report, 
                            roc_auc_score, roc_curve, auc)
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def calculate_metrics(y_true, y_pred, method_name):
    print(f"\n{'='*50}")
    print(f"МЕТРИКИ КАЧЕСТВА: {method_name}")
    print(f"{'='*50}")
    
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)
    
    print(f"Точность (Accuracy): {accuracy:.4f}")
    print(f"Точность (Precision): {precision:.4f}")
    print(f"Полнота (Recall): {recall:.4f}")
    print(f"F1-мера: {f1:.4f}")
    
    print(f"\nДетальные метрики по классам:")
    print(classification_report(y_true, y_pred, target_names=target_names, zero_division=0))
    
    return {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1
    }

metrics_euclidean = calculate_metrics(y_test, y_pred_euclidean, "Евклидово расстояние")
metrics_mahalanobis = calculate_metrics(y_test, y_pred_mahalanobis, "Махаланобис")
```
